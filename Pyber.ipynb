```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv

file_rides="data_ride_data.csv"
file_cities="data_city_data.csv"

rides_df = pd.read_csv(file_rides,encoding="ISO-8859-1")
cities_df = pd.read_csv(file_cities,encoding="ISO-8859-1")


pyber_df = cities_df.merge(rides_df, on="city")
pyber_df.head()
```

```python
urban_cities = pyber_df[pyber_df['type']=='Urban']
suburban_cities = pyber_df[pyber_df['type']=='Suburban']
rural_cities = pyber_df[pyber_df['type']=='Rural']

urban_ride = urban_cities.groupby(['city']).count()['ride_id'].reset_index()
urban_fare = urban_cities.groupby(['city']).mean()['fare'].reset_index()
urban_driver = urban_cities.groupby(['city']).sum()['driver_count'].reset_index()

suburban_ride = suburban_cities.groupby(['city']).count()['ride_id'].reset_index()
suburban_fare = suburban_cities.groupby(['city']).mean()['fare'].reset_index()
suburban_driver = suburban_cities.groupby(['city']).sum()['driver_count'].reset_index()

rural_ride = rural_cities.groupby(['city']).count()['ride_id'].reset_index()
rural_fare = rural_cities.groupby(['city']).mean()['fare'].reset_index()
rural_driver = rural_cities.groupby(['city']).sum()['driver_count'].reset_index()
```

```python
graph = plt.scatter(urban_driver["driver_count"],urban_fare["fare"],s=urban_ride["ride_id"]*12,linewidth=2,edgecolor='black',alpha=0.5,c="blue")
graph = plt.scatter(suburban_driver["driver_count"],suburban_fare["fare"],s=suburban_ride["ride_id"]*12,linewidth=2,edgecolor='black',alpha=0.5,c="red")
graph = plt.scatter(rural_driver["driver_count"],rural_fare["fare"],s=rural_ride["ride_id"]*12,linewidth=2,edgecolor='black',alpha=0.5,c="yellow")
plt.grid()
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('PyBer Sharing Data (2016)')
#plt.text(75,30,'Note: Circle size correlates with dirver count per city',horizontalalignment='right',verticalalignment='top')
plt.show()
```
